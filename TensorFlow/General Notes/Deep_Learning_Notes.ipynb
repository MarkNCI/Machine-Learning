```
## Gradient Descent
# Optimisation technique to minimise error over multiple steps
# On curve of a line,pick some parameter at random + measure the error it produces
# Keep doing this until you reach bottom of curve, when it rises the error increases
# Hence the term 'descent'. Marked in green below
# Don't get confused with local minima:
# if the curve flattens at one pt before the green bit, that isn't the least error bit!
```
