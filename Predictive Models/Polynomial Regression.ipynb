```python
### Polynomial Regression ###
# Data's not always linear, can have curves.
# Higher order polynomials = more complex curves
# y = mx+b(1st order polynomial), y = ax2+bx+c(2nd order), y=ax3+bx2+c(3rd order)
# Visualise the data first bs overusing curves (avoids overfitting)
```

```python
# Polynomial relationship between web page speed and the amount of purchases
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2) # Make random numbers predictable
pageSpeed = np.random.normal(3.0,1.0,1000)# Mean, std dev, size
purchaseAmount = np.random.normal(50.0,10.0,1000)/pageSpeed
plt.scatter(pageSpeed,purchaseAmount)
```

```python
# numpy polyfit = create nth degree polynomial model to minimise squared error.
# 4th order in this case.

x = np.array(pageSpeed)
y = np.array(purchaseAmount)

polyfit = np.poly1d(np.polyfit(x,y,12))# Change for + accuracy
```

```python
# Original plot with plot of predicted values
# Using polynomial for page speed times from 0-7 seconds

xp = np.linspace(0,7,100)
plt.scatter(x,y)
plt.plot(xp,polyfit(xp),c='r')
```

```python
# Measure r-squared error
from sklearn.metrics import r2_score

r2 = r2_score(y,polyfit(x))
r2
```
