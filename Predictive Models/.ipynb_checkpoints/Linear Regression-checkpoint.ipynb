```python
### Linear Regression ###
# Fit line to dataset of observations to predict unobserved values.
# Use least squares: minimise the squared error between each pt and line.
# R-Squared: 0 is bad, 1 is good
```

```python
# Linear relationship between web page speed and the amount of purchases
import numpy as np
import matplotlib.pyplot as plt

pageSpeeds = np.random.normal(3.0,1.0,1000)# Mean, std dev, size
purchaseAmount = 100-(pageSpeeds+np.random.normal(0,0.1,1000))*3
plt.scatter(pageSpeeds, purchaseAmount)
```

```python
# Use scipy to extract below values and get R Squared value
from scipy import stats

slope, intercept, r_value, p_value, std_err = stats.linregress(pageSpeeds, purchaseAmount)
r_value ** 2
```

```python
# Visulise line through data
def prediction(x):
    return slope * x + intercept

fitline = prediction(pageSpeeds)
plt.scatter(pageSpeeds,purchaseAmount)
plt.plot(pageSpeeds, fitline, c='r')
```
