```python
## Unsupervised Learning ##

# Good for unlabelled data
# Model trained on observations of dataset vs correct answers
# If clustering similar objects, model trained on characteristics (colour, shape)
# Used for latent variables (don't know what I'm looking for)
# Clustering users on a dating site based on info/behaviour vs known stereotypes
# Or cluster product desc text for terms related to certain categories
```

```python
## Supervised Learning ##

# Good for labelled data
# Model trained on correct answers to predict new/unknown values
# Evaluate using train/test: Train on 80% of dataset, test on remaining 20%
# Good way to prevent overfitting
# Datasets must be randomly selected + large enough to represent all variants/outliers
# But overfitting can still happen (k-fold cross validation)
```

```python
# Polynomial Regression with Train/Test
import numpy as np
import matplotlib.pyplot as plt

np.random.seed(2)
pageSpeeds = np.random.normal(3.0,1.0,100)
purchaseAmount = np.random.normal(50.0,30.0,100)/pageSpeeds
plt.scatter(pageSpeeds, purchaseAmount)
```

```python
# Implement Train/Test: 80/20 split (Real World: random.shuffle data first)
trainX = pageSpeeds[:80]
testX = pageSpeeds[20:]

trainY = purchaseAmount[:80]
testY = purchaseAmount[20:]
```

```python
# Training dataset
plt.scatter(trainX,trainY)
```

```python
# Test dataset
plt.scatter(testX,testY)
```

```python
# 8th Degree Polynomial as it'll overfit
x = np.array(trainX)
y = np.array(trainY)

p4 = np.poly1d(np.polyfit(x,y,8))
```

```python
# Training data plot with r-squared error
from sklearn.metrics import r2_score

xp = np.linspace(0,7,100)
axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])

plt.scatter(x,y)
plt.plot(xp,p4(xp),c='r')
plt.show()

r2 = r2_score(np.array(trainY),p4(np.array(trainX)))
r2

# Below graph is overfitting at end
```

```python
# Test data plot with r-squared error

testx = np.array(testX)
testy = np.array(testY)

axes = plt.axes()
axes.set_xlim([0,7])
axes.set_ylim([0,200])

plt.scatter(testx,testy)
plt.plot(xp,p4(xp),c='r')
plt.show()

r2 = r2_score(testy,p4(testx))
r2
```
