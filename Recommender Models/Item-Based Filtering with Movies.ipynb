```python
# Load pickle from User based model
import pickle
with open(r'C:\Users\MHayden\Desktop\Machine Learning Library\Recommender Models\movieRatings','rb') as userRatings:
    userRatings = pickle.load(userRatings)
    
userRatings.head()
```

```python
# Pandas corr() = compute correlation for every movie pair where at least 1 user rated both movies
userMatrix = userRatings.corr()
userMatrix.head()
```

```python
# Like before, results skewed when small number of users have watched same pair of movies (2 people only watched 2 films for example)
# So use min_periods ro remove ratings with < 100 users (Use pearson correlation method here)
viewerMatrix = userRatings.corr(method='pearson',min_periods=100)
viewerMatrix.head()
```

```python
# Recommend movies for user_id 0
testRating = userRatings.loc[0].dropna()
testRating
```
